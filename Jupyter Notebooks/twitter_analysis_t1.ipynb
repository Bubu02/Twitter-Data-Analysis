#### ***1. Plot a scatter chart to analyse the relationship between media engagements and media views for tweets that received more than 10 replies. Highlight tweets with an engagement rate above 5% and this graph should work only between 6PM IST to 11 PM IST apart from that time we should not show this graph in dashboard itself and the tweet date should be odd number as well as tweet*** letter count ***be above 50.***

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('twitter.csv')
```

```python
df.head()
```

```python
# Convert 'time' to datetime objects
df['time'] = pd.to_datetime(df['time'])
```

```python
df['time'].head()
```

```python
# Filter 1: More than 10 replies
filtered_df = df[df['replies'] > 10].copy()
```

```python
filtered_df.head(5)
```

```python
df.info()
```

```python
filtered_df.info()
```

```python
# Filter 2: Tweet date is an odd number
filtered_df['day'] = filtered_df['time'].dt.day
filtered_df = filtered_df[filtered_df['day'] % 2 != 0]
```

```python
filtered_df.info()
```

```python
# Filter 2: Tweet date is an odd number
filtered_df['day'] = filtered_df['time'].dt.day
filtered_df = filtered_df[filtered_df['day'] % 2 != 0]
```

```python
filtered_df.info()
```

```python
# Filter 3: Tweet word count be above 50
filtered_df['word_count'] = df['Tweet'].str.split().str.len()
filtered_df = filtered_df[filtered_df['word_count'] > 50]
```

```python
filtered_df.info()
```

```python
# Calculate engagement rate
filtered_df['engagement_rate'] = (filtered_df['media engagements'] / filtered_df['media views']) * 100

# Separate tweets with engagement rate > 5%
highlight_df = filtered_df[filtered_df['engagement_rate'] > 5]
```

```python
# Create the scatter plot
plt.figure(figsize=(12, 8))
plt.scatter(filtered_df['media views'], filtered_df['media engagements'], label='Engagement Rate <= 5%')
plt.scatter(highlight_df['media views'], highlight_df['media engagements'], color='red', label='Engagement Rate > 5%')

# Add titles and labels
plt.title('Media Engagements vs. Media Views')
plt.xlabel('Media Views')
plt.ylabel('Media Engagements')
plt.legend()
plt.grid(True)

# Add description of filters to the plot
filter_description = "Filters Applied:\n- Replies > 10\n- Tweet date is an odd number\n- Word count > 50"
plt.text(0.05, 0.95, filter_description, transform=plt.gca().transAxes, fontsize=10,
         verticalalignment='top', bbox=dict(boxstyle='round,pad=0.5', fc='wheat', alpha=0.5))

# Save the plot to a file
# plt.savefig('media_engagement_plot.png')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set Seaborn style
sns.set(style="whitegrid")

# Create a new column to label engagement rate category
filtered_df_02['engagement_category'] = 'Engagement Rate <= 5%'
highlight_df_02.loc[:, 'engagement_category'] = 'Engagement Rate > 5%'

# Combine both DataFrames for plotting
combined_df = pd.concat([filtered_df_02, highlight_df_02])

# Create the scatter plot
plt.figure(figsize=(12, 8))
sns.scatterplot(
    data=combined_df,
    x='media views',
    y='media engagements',
    hue='engagement_category',
    palette={'Engagement Rate <= 5%': 'blue', 'Engagement Rate > 5%': 'red'},
    s=100,  # marker size
    alpha=0.7
)

# Add titles and labels
plt.title('Media Engagements vs. Media Views', fontsize=16)
plt.xlabel('Media Views', fontsize=12)
plt.ylabel('Media Engagements', fontsize=12)
plt.legend(title='Engagement Category')
plt.grid(True)

# Add description of filters to the plot
filter_description_02 = (
    "Filters Applied:\n"
    "- Replies > 10\n"
    "- Tweet date is an odd number\n"
    "- Word count > 50"
)
plt.text(
    0.05, 0.95, filter_description_02,
    transform=plt.gca().transAxes,
    fontsize=10,
    verticalalignment='top',
    bbox=dict(boxstyle='round,pad=0.5', fc='wheat', alpha=0.5)
)

# Save and show the plot
plt.savefig('media_engagement_plot_seaborn.png')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime
import pytz
import pandas as pd

# Set Seaborn style
sns.set(style="whitegrid")

# Create a new column to label engagement rate category
filtered_df_02['engagement_category'] = 'Engagement Rate <= 5%'
highlight_df_02.loc[:, 'engagement_category'] = 'Engagement Rate > 5%'

# Combine both DataFrames for plotting
combined_df = pd.concat([filtered_df_02, highlight_df_02])

# Create the scatter plot
plt.figure(figsize=(12, 8))
sns.scatterplot(
    data=combined_df,
    x='media views',
    y='media engagements',
    hue='engagement_category',
    palette={'Engagement Rate <= 5%': 'blue', 'Engagement Rate > 5%': 'red'},
    s=100,  # marker size
    alpha=0.7
)

# Add titles and labels
plt.title('Media Engagements vs. Media Views', fontsize=16)
plt.xlabel('Media Views', fontsize=12)
plt.ylabel('Media Engagements', fontsize=12)
plt.legend(title='Engagement Category')
plt.grid(True)

# Add description of filters to the plot
filter_description_02 = (
    "Filters Applied:\n"
    "- Replies > 10\n"
    "- Tweet date is an odd number\n"
    "- Word count > 50"
)
plt.text(
    0.05, 0.95, filter_description_02,
    transform=plt.gca().transAxes,
    fontsize=10,
    verticalalignment='top',
    bbox=dict(boxstyle='round,pad=0.5', fc='wheat', alpha=0.5)
)

# --- Add visual timestamp annotation (IST) ---
ist = pytz.timezone('Asia/Kolkata')
current_time_ist = datetime.now(ist).strftime("%Y-%m-%d %I:%M:%S %p %Z")

plt.text(
    0.95, 0.02,  # bottom-right corner
    f"Generated: {current_time_ist}",
    transform=plt.gca().transAxes,
    fontsize=9,
    color='gray',
    ha='right',
    va='bottom',
    bbox=dict(boxstyle='round,pad=0.3', fc='white', ec='gray', alpha=0.7)
)

# Save and show the plot
plt.tight_layout()
plt.savefig('media_engagement_plot_seaborn_timestamped.png', dpi=300)
plt.show()
```
